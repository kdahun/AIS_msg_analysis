```python
import pandas as pd

FILE_PATH = 'AIS_부산 실해역 테스트 데이터3.xlsx'
```

```python
# 전체 시트 읽기 (헤더 없음)
sheets = pd.read_excel(FILE_PATH, sheet_name=None, header=None)

for name, df in sheets.items():
    print(f'\n=== 시트: {name} ===')
    print(f'행: {df.shape[0]}, 열: {df.shape[1]}')
    print('샘플:', df.iloc[0].tolist())
```
